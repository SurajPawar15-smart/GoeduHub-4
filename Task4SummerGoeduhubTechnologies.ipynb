###***Q.18 Create the following pattern without hardcoding. Use only numpy functions and the below input array a.***<br>
Input:<br>
a = np.array([1,2,3])<br>
Desired Output:<br>
array([1, 1, 1, 2, 2, 2, 3, 3, 3, 1, 2, 3, 1, 2, 3, 1, 2, 3])

```python
a=np.array([1,2,3])
print(a)
a=np.append(np.repeat(a,3),np.tile(a,3))
print(a)
```
